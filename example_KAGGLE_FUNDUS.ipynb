# Illustration of data from Kaggle retina fundus dataset

https://www.kaggle.com/jr2ngb/cataractdataset/version/2
stored in data/KaggleFundus

Here is 2D retina fundus image of a retina from a healthy patient with cataract, glaucoma and retinal disease. 
This notebook is a quick example of how tha data look like. 

```python
import os
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
dirname = os.getcwd()
extension=".png"
```

```python

fig,ax=plt.subplots(1,4,figsize=(20,5))

filepathname_normal=os.path.join(dirname, 'data','KaggleFundus','NL_001'+extension)
filepathname_cataract=os.path.join(dirname, 'data','KaggleFundus','cataract_003'+extension)
filepathname_glaucoma=os.path.join(dirname, 'data','KaggleFundus','Glaucoma_002'+extension)
filepathname_retina=os.path.join(dirname, 'data','KaggleFundus','Retina_001'+extension)

retina_normal = mpimg.imread(filepathname_normal)
retina_cataract = mpimg.imread(filepathname_cataract)
retina_glaucoma = mpimg.imread(filepathname_glaucoma)
retina_retina = mpimg.imread(filepathname_retina)

ax[0].imshow(retina_normal)
ax[0].set_title("Healthy cornea ")
ax[1].imshow(retina_cataract)
ax[1].set_title("Cataract")
ax[2].imshow(retina_glaucoma)
ax[2].set_title("Glaucoma")
ax[3].imshow(retina_retina)
ax[3].set_title("Retinal disease")
```

Some observation: 
- In cataract images seems blury 
- Retinal disease: texture different
