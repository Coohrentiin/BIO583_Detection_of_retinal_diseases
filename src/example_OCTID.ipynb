# Example of data from OCTID dataset

```python
import os
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
dirname = os.getcwd()
dirname=os.path.join(dirname,"..")
extension=".jpeg"
%matplotlib inline
print(dirname)
```

```python
N=5
fig,ax=plt.subplots(N,2,figsize=(10,18))
for n in range(1,N+1):
	filepathname_normal=os.path.join(dirname, 'data','OCTID','normal','NORMAL'+str(n)+extension)
	retina_normal = mpimg.imread(filepathname_normal)
	filepathname_amd=os.path.join(dirname, 'data','OCTID','amd','AMRD'+str(n)+extension)
	retina_amd = mpimg.imread(filepathname_amd)
	ax[n-1][0].imshow(retina_normal)
	ax[n-1][0].set_title("Healthy cornea "+str(n))
	ax[n-1][1].imshow(retina_amd)
	ax[n-1][1].set_title("Cornea with Amd "+str(n))
```
